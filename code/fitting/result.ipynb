```python
def keep_last_two(path):
        parts = os.path.normpath(path).split(os.sep)
        return "/".join(parts[-3:])

def find_best_sample(csv_path, beta=1.0, gamma=1.0, match_rate_thresh=0.5, top_n=8):
    
    df = pd.read_csv(csv_path)

    df_filtered = df[df["match_rate"] > match_rate_thresh].copy()

    df_filtered["Score"] = beta * df_filtered["rmsd"] + gamma * df_filtered["mean_shift"].abs()

    df_top = df_filtered.nsmallest(top_n, "Score")

    first_col = df_filtered.columns[0]
    df_top[first_col] = df_top[first_col].apply(keep_last_two)

    result = df_top[[first_col, "Score"]].reset_index(drop=True)

    print(result)
    return result
```

```python
best_name, best_score = find_best_sample(
    "/mnt/data/SYC/xasfitting/LaCe_L2_peak_fitting_results.csv",
    beta=0.6, gamma=0.4
)
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

img_dir = "/mnt/data/SYC/xasfitting/1112/pic" 
save_path = os.path.join(img_dir, "Ce-L3_param4.png")

files = sorted([f for f in os.listdir(img_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

cols = 3
rows = 4
figsize = (25, 28) 

fig, axes = plt.subplots(rows, cols, figsize=figsize)
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < len(files):
        img_path = os.path.join(img_dir, files[i])
        img = Image.open(img_path)
        ax.imshow(img)
        ax.axis('off')
    else:
        ax.axis('off')

plt.tight_layout(pad=0.5, w_pad=0.3, h_pad=0.5)
plt.savefig(save_path, dpi=300)
plt.show()

print(f"✅  {save_path}")
```
