```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_xy(path):
    d = np.loadtxt(path)
    return d[:,0], d[:,1]
```

CoCe_Ce_L3

```python
x1, y1 = load_xy('/mnt/data/SYC/xasfitting/proced/CoCe_Ce_L3_merged.txt')
plt.figure(figsize=(10,5))
plt.plot(x1, y1)

plt.title("CoCe_Ce_L3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.tight_layout()
plt.show()
```
